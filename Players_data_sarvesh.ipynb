We need to define new features for both batsman and bowlers data
list of new features is as follows :

1.Batsman:
hard_hitter, finisher, fast_scorer, consistent, run_btn_wkt

2.Bowlers:
Economy(which is already there), wicket_taker, consistent, big_wicket_taker, short_performances
for corresponding expressions and formulas of the features refer to this paper
https://pdfs.semanticscholar.org/5314/b20bbc2faf3069714dc69abdb35c00a68f6f.pdf


```python
batsman['Inn'] = batsman['Inn'].apply(lambda x:float(x))
batsman['NO'] = batsman['NO'].apply(lambda x:float(x))
batsman['4s'] = batsman['4s'].apply(lambda x:float(x))
batsman['6s'] = batsman['6s'].apply(lambda x:float(x))
batsman['Balls'] = batsman['Balls'].apply(lambda x:float(x))
batsman['Runs'] = batsman['Runs'].apply(lambda x:float(x))
```

```python
batsman['hard_hitter'] = (4*batsman['4s'] + 6*batsman['6s'])/(batsman['Balls'])
```

```python
batsman['finisher'] = (batsman['NO']/batsman['Inn'])
```

```python
batsman['fast_scorer'] = (batsman['Runs']/batsman['Balls'])
```

```python
batsman['consistent'] = (batsman['Runs']/(batsman['Inn']-batsman['NO']))
```

```python
batsman['balls_without_boundry'] = batsman['Balls'] - (batsman['4s']+batsman['6s'])
```

```python
batsman['run_betn_wkts'] = (batsman['Runs']-(4*batsman['4s'] + 6*batsman['6s']))/batsman['balls_without_boundry']
```

```python
bowler
```

```python
bowler['Wkt_Taker'] = (bowler['Balls']/bowler['Wkts'])
```

```python
bowler['consistent'] =  (bowler['Runs']/bowler['Wkts'])
```

```python
bowler['big_wkt_taker'] = (bowler['4w']+bowler['5w'])/bowler['Inn']
```

```python
bowler['short_performance'] = (bowler['Wkts']-(bowler['4w']+bowler['5w']))/(bowler['Inn']-(bowler['4w']+bowler['5w']))
```
